## Identificación valores nulos
* Se aprecia que no existen valores nulos, sin embargo, denotamos que en el dataframe existen datos con el signo de interrogación '?'.
* Esto nos llevo a interpretar que asi se representan los valores nulos.
A continuación, se evidencia el proceso de cambiar esos valores nulos '?' de 3 columnas las cuales son:
1. clase_trabajo: 2799 datos con el signo '?'

3. pais: 857 datos con el signo '?'

# Análisis Diagrama de Cajas y Bigotes
1. Se determinará la distribucion de datos con la columna num_educacion, dado que es aquella con un rango adecuado de datos para contrastar e identificar a que pais pertenecera el dato nulo '?'.

```python
plt.figure(figsize=(12, 8)) 
sns.boxplot(x="num_educacion", y="pais", data=data_eu2)
plt.yticks(rotation=45)
# Mostrar el gráfico
plt.show()
```

```python
def asignar_pais_segun_num_educacion(row, pais_num_educacion):
    if row['pais'] == '?':
        return pais_num_educacion.get(row['num_educacion'], 'Desconocido')
    return row['pais']
pais_num_educacion = data_eu2[data_eu2['pais'] != '?'].groupby('num_educacion')['pais'].agg(lambda x: x.value_counts().index[0]).to_dict()
data_eu2['pais'] = data_eu2.apply(asignar_pais_segun_num_educacion, axis=1, pais_num_educacion=pais_num_educacion)
```

## Transformación variables categoricas a númericas.

```python
data_eu2.head()
```

```python
sexo = pd.get_dummies(data_eu2["sexo"], drop_first=True)
#se añade al dataset y borro la que ya no hace falta
data_eu2 = pd.concat([data_eu2,sexo],axis=1)
data_eu2.drop("sexo",axis=1,inplace=True)
data_eu2.head(5)
```

```python
data_eu2=data_eu2.rename(columns={"Male":"sexo"})
data_eu2.head(3)
```

```python
#TRANSFORMAR COLUMNA clase_trabajo
data_eu2['clase_trabajo'] = pd.Categorical(data_eu2['clase_trabajo']).codes + 1
#TRANSFORMAR COLUMNA educacion
data_eu2['educacion'] = pd.Categorical(data_eu2['educacion']).codes + 1
#TRANSFORMAR COLUMNA estado_civil
data_eu2['estado_civil'] = pd.Categorical(data_eu2['estado_civil']).codes + 1
#TRANSFORMAR COLUMNA ocupacion
data_eu2['ocupacion'] = pd.Categorical(data_eu2['ocupacion']).codes + 1
#TRANSFORMAR COLUMNA relacion
data_eu2['relacion'] = pd.Categorical(data_eu2['relacion']).codes + 1
#TRANSFORMAR COLUMNA raza
data_eu2['raza'] = pd.Categorical(data_eu2['raza']).codes + 1
#TRANSFORMAR COLUMNA pais
data_eu2['pais'] = pd.Categorical(data_eu2['pais']).codes + 1
```
