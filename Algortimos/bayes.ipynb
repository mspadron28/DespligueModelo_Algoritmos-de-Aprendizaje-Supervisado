```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.naive_bayes import GaussianNB
```

```python
data = pd.read_csv('data_evaluacion_limpia_categorical.csv')
data.drop("Unnamed: 0",axis=1,inplace=True)
data
```

```python
x = data.drop(columns=['salario']).values
y = data['salario'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# Evaluacion del modelo
print('Train Accuracy :', (model.score(X_train, y_train))*100,"%")
print('Test Accuracy :', (model.score(X_test, y_test))*100,"%")
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

## Despliegue del modelo

```python
import pickle
pkl_filename = "ingreso_model.pkl"
with open(pkl_filename, 'wb') as file:
    pickle.dump(model, file)
```
